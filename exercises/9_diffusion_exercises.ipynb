### Supporting graphic for exercise 9.13

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensions to explore
dims = [2, 3, 4, 8, 16, 32, 128, 512, 768]
N = 5000  # number of random comparison vectors

fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.flatten()

for i, d in enumerate(dims):
    # Reference vector
    ref = np.random.randn(d)
    ref = ref / np.linalg.norm(ref)

    # Random vectors
    X = np.random.randn(N, d)
    X = X / np.linalg.norm(X, axis=1, keepdims=True)

    # Cosine similarities
    cosines = X @ ref

    ax = axes[i]
    ax.hist(cosines, bins=100)
    ax.set_title(f"d = {d}")
    ax.set_xlim([-1, 1])
    plt.savefig('/Users/stephen/Stephencwelch Dropbox/welch_labs/ai_book/9_sora/exercises/9_13.svg')
```
